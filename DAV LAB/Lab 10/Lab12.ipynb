```python
df = pd.read_csv('Google_data_cleaned.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
data = df["content_rating"].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(data.index, data.values)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(data.values, labels=data.index, autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df["size(kb)"], bins=20)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["rating"], df["reviews"], cmap="viridis")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(
    x="update_year",
    y="reviews",
    showmeans=True,
    df=df,
)

plt.show()
```
