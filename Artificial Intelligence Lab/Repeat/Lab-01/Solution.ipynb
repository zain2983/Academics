```python
def reverse_string(s):
    left, right = 0, len(s) - 1
    while left < right:
        # Swap characters
        s[left], s[right] = s[right], s[left]
        # Move pointers
        left += 1
        right -= 1

s1 = ["h", "e", "l", "l", "o"]
reverse_string(s1)
print(s1)

s2 = ["H", "a", "n", "n", "a", "h"]
reverse_string(s2)
print(s2)
```

```python
def is_palindrome(s):
    cleaned = ''.join(char.lower() for char in s if char.isalnum())
    return cleaned == cleaned[::-1]

s1 = "A man, a plan, a canal: Panama"
print(is_palindrome(s1))

s2 = "race a car"
print(is_palindrome(s2))

s2 = "1arra1"
print(is_palindrome(s2))
```

```python
def sqrt(x):
    if x < 2:
        return x
    
    left, right = 0, x
    while left <= right:
        mid = (left + right) // 2
        if mid * mid == x:
            return mid
        elif mid * mid < x:
            left = mid + 1
        else:
            right = mid - 1
    
    return right

x1 = 4
print(sqrt(x1))

x2 = 64
print(sqrt(x2))

x2 = 225
print(sqrt(x2))
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
```

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({i: species for i, species in enumerate(iris.target_names)})

print(df.head())
```

```python
print(df.shape , "\n")
print(df.columns)
```

```python
# Group by species and calculate the mean of each feature
df_mean = df.groupby('species').mean()

# Plot a bar graph
df_mean.plot(kind='bar', figsize=(10, 6))
plt.title('Mean of Each Feature by Species')
plt.xlabel('Species')
plt.ylabel('Mean Values')
plt.show()
```

```python
df_mean.plot(kind='line', figsize=(10, 6))
plt.title('Mean of Each Feature by Species')
plt.xlabel('Species')
plt.ylabel('Mean Values')
plt.xticks(ticks=range(3), labels=iris.target_names, rotation=0)
plt.legend(loc='upper left')
plt.show()
```

```python
species_count = df['species'].value_counts()
species_count.plot(kind='pie', labels=iris.target_names, autopct='%1.1f%%', figsize=(8, 8))
plt.title('Distribution of Species in the Iris Dataset')
plt.show()
```

```python
df['species'] = iris.target

colors = ['red', 'green', 'blue']
species_names = iris.target_names

plt.figure(figsize=(10, 6))
for species_id, color in enumerate(colors):
    plt.scatter(df[df['species'] == species_id]['sepal length (cm)'],
                df[df['species'] == species_id]['sepal width (cm)'],
                color=color, label=species_names[species_id])

plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
import csv

data = [
    ['Name', 'Age', 'Height','GPA'],
    ['Ibrahim', '30', "5.10",'2.3'],
    ['Tahir', '25', "6.0",'3.0'],
    ['Musa', '35', "5.8",'3.5']
]

filename = "output.csv"

with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print(f"'{filename}' created successfully.")
```

```python
ages = []
heights = []
gpas = []
names = []

filename = "output.csv"

with open(filename, mode='r') as file:
    reader = csv.reader(file)
    next(reader) 
    for row in reader:
        names.append(row[0])
        ages.append(int(row[1]))
        heights.append(float(row[2]))
        gpas.append(float(row[3]))
```

```python
plt.figure(figsize=(10, 5))
plt.bar(names, ages)
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Bar Plot for Age')
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(names, ages)
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Line Graph of Age')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(names, heights)
plt.xlabel('Names')
plt.ylabel('Height')
plt.title('Bar Plot for Height')
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(names, heights)
plt.xlabel('Names')
plt.ylabel('Heights')
plt.title('Line Graph of Heights')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(names, gpas)
plt.xlabel('Names')
plt.ylabel('GPA')
plt.title('Bar Plot for GPA')
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(names, gpas)
plt.xlabel('Names')
plt.ylabel('GPA')
plt.title('Line Graph of GPA')
plt.show()
```
